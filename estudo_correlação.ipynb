# Estudo Correlação

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/dataset-saude-mental.tsv', sep='\t')
```

```python
df.head()
```

```python
df.columns
```

## Somando as variáveis do GAD

```python
df['gad'] = df['gad-1'] + df['gad-2'] + df['gad-3'] + df['gad-4'] + df['gad-5'] + df['gad-6'] + df['gad-7']
df.head()
```

## Estudando a correlação entre variáveis intervalares

```python
ndf1 = df.filter(['idade', 'renda', 'gad', 'motivação'])
ndf1.head()
```

### Pearson

```python
df.corr(method="pearson")
```

```python
ndf1.corr(method='pearson')
```

### Spearman

```python
ndf1.corr(method='spearman')
```

## Estudando a correlação entre variáveis nominais e ordinais

```python
ndf2 = df.filter(['cor', 'sexo', 'escola', 'cod_curso', 'motivação'])
ndf2.head()
```

```python
!pip install researchpy
```

```python
import researchpy as rp

table, results = rp.crosstab(ndf2['cor'], ndf2['motivação'], test='chi-square')
print(results)
```

```python
table, results = rp.crosstab(ndf2['sexo'], ndf2['motivação'], test='chi-square')
print(results)
```

```python
table, results = rp.crosstab(ndf2['cod_curso'], ndf2['motivação'], test='chi-square')
print(results)
```
